```python
#import libraries to visualize dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#scatterplot to show relationship between Sepal Length and Sepal Width coloured by species
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.legend(title='species',loc='upper left', bbox_to_anchor=(1.0, 1))
plt.show()
```

```python
#scatterplot to show relationship between petal Length and petal Width coloured by species
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.title('Scatter Plot of petal Length vs petal Width')
plt.legend(title='species',loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
# Pairplot to show relationships between all feature pairs, coloured by species
sns.pairplot(df, hue="species")
plt.suptitle("Feature Relationships by Species", y=1)
plt.show()
```

```python
#set theme to visualize histogram better
sns.set_theme(style='whitegrid', palette='Blues')
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
# Histograms Value Distributions
df.hist(figsize=(8, 8), bins=10)
plt.suptitle("Feature Distributions", y=1.02)
plt.tight_layout()
plt.show()
```

Box Plot For Outlier detection

```python
# Identify numerical columns for individual box plots
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Create a figure with subplots for each numerical column
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1) # Arrange plots in a 2x2 grid
    sns.boxplot(x=df[col]) # Plot each column on the y-axis
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col) # Label the y-axis with the column name

plt.suptitle("Box Plots for Numerical Features to Identify Outliers", y=1.02)
plt.tight_layout()
plt.show()
```
